```python
def num_poi(col1, col2):

    url = "https://api.foursquare.com/v3/places/search"

    params = {"query": "supermarket",
  	    "ll": f"{col1},{col2}",
        "radius": 10000,
  	    "sort":"DISTANCE"}

    headers = {"Accept": "application/json",
                "Authorization": api_key}
    try:
        response = requests.request("GET", url, params=params, headers=headers)
        dict = response.json()
        dic2 = dict["results"]
        num = len(dic2)
        return num
    except ConnectionError:
        time.sleep(15)
```

```python
df["n_supermarket_10km"] = df.apply(lambda pepe: num_poi(pepe["latitude"], pepe["longitude"]), axis=1)
```
